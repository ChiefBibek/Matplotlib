```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7]
y = [ 50,51,52,48,47,49,46]
```

```python
plt.plot(x,y,color='green')
```

```python
plt.plot(x,y,'bo')
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,color='green', linewidth=5,linestyle='dotted')
```

Format Strings In Plot Functions

```python
plt.plot (x,y,'g+')
```

```python
plt.plot (x,y,'g+--')
```

```python
plt.plot (x,y,'g+-.')
#here g means green 
# we cane change it to red as r too
```

```python
plt.plot(x,y,color="#00ffff",marker="*",linestyle= '--', markersize= 20)
```

```python
plt.plot(x,y,alpha=0.2)
```
